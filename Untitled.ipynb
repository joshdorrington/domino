```python
import xarray as xr
e=xr.open_dataarray('examples/data/NorthItalian_SON_90Percentile_5dPrecip.nc')
Z=xr.open_dataarray('examples/data/Z500.nc')
regimes=xr.open_dataarray('examples/data/Grams_regime_lifecycles.nc')
```

```python
a=LaggedAnalyser(event=e,variables=Z)
```

```python
a.add_variable(regimes,name='Regimes',is_categorical=True,join_type='left')
```

```python
a.lag_variables(range(-2,3))
```

```python
a.lagged_variables(-1)
```

```python
average_vals=a.aggregate_variables()
```

```python
C=a.compute_composites()
```

```python
merid_deriv=lambda da: da.differentiate('lat')
a.add_derived_composite('dxZ',merid_deriv,['Z'])
```

```python
S=a.bootstrap_significance(100,0.05)
```
